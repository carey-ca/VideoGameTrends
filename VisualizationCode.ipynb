```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rush4ratio/video-game-sales-with-ratings")

print("Path to dataset files:", path)
```

```python
import pandas as pd
vg_sales = pd.read_csv(f"{path}/Video_Games_Sales_as_at_22_Dec_2016.csv")

vg_sales['User_Score'] = pd.to_numeric(vg_sales['User_Score'], errors='coerce')
vg_sales['Critic_Score'] = pd.to_numeric(vg_sales['Critic_Score'], errors='coerce')

vgs_cleaned = vg_sales.dropna(subset=['User_Score', 'Critic_Score'])

vgs_cleaned
```

```python
import altair as alt
alt.data_transformers.enable('default', max_rows=100000)

alt.Chart(vgs_cleaned).transform_calculate(
    User_Score_100='datum.User_Score * 10'
).mark_circle(
    size=20,
    ).encode(
    alt.Y('User_Score_100:Q', title='User Score (0-100)'),
    alt.X('Critic_Score:Q', title='Critic Score (0-100)'),
    alt.Color('Platform:N'),
    alt.Size('Global_Sales:Q', scale=alt.Scale(range=[10, 500])),
    tooltip=['Name', 'Year_of_Release', 'Genre', 'Global_Sales', 'Critic_Count', 'User_Count', 'Developer', 'Rating']
).properties(
    width=1000,
    height=800,
    title='User Score vs Critic Score by Platform and Rating'
)
```
